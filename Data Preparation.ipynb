<h3>Merge two table</h3>

Step 5:
Next, combine the two table using pd.merge. pd.merge(the first table, the second table to combine, on is using the similarity attribute from both table to combine. combine_table.head() to show the five above dataset to see the combine_table.

```python
combine_table= pd.merge(customer,product2,on='customer_id')
combine_table.head()
```

select certain attribute such as age and output into age_subscribe. This information to explorer range of age that subscribe the term deposit either yes or no. using the groupby to group the output and get the sum of age that for the category fall under 'no' and 'yes'

```python
age_subscribe=combine_table[['age', 'output']]
age_subscribe.head()
```

```python
grouped= age_subscribe.groupby('output')

print(grouped['age'].count())
```

Step 6:
Create frequency tables (also known as crosstabs) in pandas using the pd.crosstab() function. The function takes one or more array-like objects as indexes or columns and then constructs a new DataFrame of variable counts based on the supplied arrays.The marginal counts (totals for each row and column) by including the argument margins=True.

```python
age_tab = pd.crosstab(index=age_subscribe['age'],columns=age_subscribe['output'], margins=True)

age_tab
```

describe() is to check the basic statistic information for age.

```python
age_subscribe['age'].describe()
```

import the numpy. use numpy to check the median of age.

```python
import numpy as np
np.median(age_subscribe['age'])
```

based on above information, the age can be store into 3 category according to min, mean and max. 
pd.cut is use to cut when to segment and sort data values into bins. This function is also useful for going from a continuous variable to a categorical variable. For example, cut could convert ages to groups of age ranges. Supports binning into an equal number of bins, or a pre-specified array of bins.
include_lowest is bool. Whether the first interval should be left-inclusive or no.
The age category to be stored in age_subscribe['new_age']

```python
age_subscribe['new_age'] = pd.cut(age_subscribe.age, 
                   bins=[18,33,45,95], 
                   labels=['Young','Young Adult','Adult'],include_lowest=True)
age_subscribe.head()
```
